```python
from bs4 import BeautifulSoup
import requests
 
def scrape(URL):
    # opening our output file in append mode
    File = open("productdata1.csv", "a")
 
    # specifying user agent, You can use other user agents
    # available on the internet
    HEADERS = ({'User-Agent':
                'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
                                'Accept-Language': 'en-US, en;q=0.5'})
 
    # Making the HTTP Request
    webpage = requests.get(URL, headers=HEADERS)
 
    # Creating the Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "lxml")
    
 
    #retrieving product title
    try:
        # Outer Tag Object
        title = soup.find("span",
                          attrs={"id": 'productTitle'})
 
        # Inner NavigableString Object
        title_value = title.string
 
        # Title as a string value
        title_string = title_value.strip().replace(',', '')
 
    except AttributeError:
        title_string = "NA"
    print("product Title = ", title_string)
    File.write(f"{title_string},")
    
    #ASIN
    try:
        # Outer Tag Object
        asin = soup.find("div",
                          attrs={"id": 'detailBullets_feature_div'})
        
        asin = asin.find_all("span")
        
        asin = str(asin[len(asin)-1]).split('<span>')
        asin = str(asin[1]).split('<')
        asin = asin[0]
 
    except AttributeError:
        asin = "NA"
    print("ASIN = ", asin)
    File.write(f"{asin},")
    
    #Manufacturer/Brand
    try:
        # Outer Tag Object
        brand = soup.find("a",
                          attrs={"id": 'bylineInfo'})
        brand = str(brand).split('Visit the')
        brand = brand[len(brand)-1].split('Store')
        brand = brand[0]
        #brand = brand.find_all('li')
        #brand = str(brand).split('Manufacturer\n                                    \u200f\n                                        :\n                                    \u200e\n                                </span> <span>')
        #brand = brand[len(brand)-1].split('</span> </span>')
        #brand = brand[0]
        
        #if '<' in brand:
         #   brand = "NA"
        
        
    except AttributeError:
        brand = "NA"
    print("Brand =", brand)
    File.write(f"{brand},")
    
    #Primary Material
    try:
        material = soup.find("div",
                          attrs={"id": 'feature-bullets'})
        material = material.find('li')
        material = str(material).split("%")
        try:
            material = material[1]
            material = material.split(',')
            material = material[0].strip()
        except IndexError:
            material = 'NA'
            
    except AttributeError:
        material = "NA"
        
    print("Material =", material)
    File.write(f"{material},")
    
    #Country aka Imported
    
    try:
        country = soup.find("div",
                          attrs={"id": 'feature-bullets'})
        country = country.find_all('li')
        if 'Imported' in str(country):
            country = "Imported"
        else:
            country = "Not Imported"
    except AttributeError:
        country = 'Not Imported'
    
    print("Country =", country)
    File.write(f"{country}\n,")
    File.close()
    
if __name__ == '__main__':
  # opening our url file to access URLs
    FILE = open("urlsscrape.txt", "r")
    for link in FILE:
        scrape(link)
    
```
